```python
import numpy as np
from matplotlib import pyplot as plt
import os
from circuits_ttt import *
plt.rcParams.update({'font.size': 14})
```

```python
path_to_analyze = './output/cluster_out/no_epochs/3'
filenames = next(os.walk(path_to_analyze), (None, None, []))[2]
symmetric = [np.load(path_to_analyze + '/' + i, allow_pickle=True).item() for i in filenames if i != 'params.txt']
with open(path_to_analyze+'/params.txt') as f:
    params = f.readlines()
print(params)
```

```python
plt.figure(figsize=(12, 8))

cost_fun = torch.tensor([i['cost function'] for i in symmetric]).detach().numpy()
print(cost_fun)
avg_cost = np.average(cost_fun, axis=0)
plt.xlabel('step')
plt.ylabel('cost')
plt.plot(avg_cost, label='avg')
for i in symmetric:
    plt.plot(torch.tensor(i['cost function'] ).detach().numpy())
plt.title(params)
plt.legend()
```
